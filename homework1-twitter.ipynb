```python
import json
import pandas
import pprint
import collections
import string
import re


twitter = []
for line in open('hw_3_twitter.json'):
    twitter.append(json.loads(line))
print(len(twitter), 'твитов')   #количество твитов, ЗАДАНИЕ 1


lang = []    #cамые популярные языки твитов
user = []    #какие пользователи повторяются
hashtag = []    #топ 20
english = []    #частотный словарь слов из твитов на анлийском
followers = []    #топ 10 по кол-ву подписчиков
ist = []    #топ 10 источников твитов


print((sum('delete' in t for t in twitter)/len(twitter))*100, '% - из них удаленных')   #процент удаленных, ЗАДАНИЕ 2


for a in twitter:
    if 'delete' in a:
        lang = lang   # надо было что-нибудь вставить, решил не заморачиваться и не искать, как красиво сделать
    else:
        lang.append(a['lang'])
counter = collections.Counter(lang) 
langs = []
for key in sorted(counter, key=counter.get):   #еще в прошлом году бесило, что получается так громоздко
    langs.append(key)                          #научите, пожалуйста, как нормально
langs.reverse()
print(langs[:5], '- топ 5 самых популярных языков' )   #нашли языки, ЗАДАНИЕ 3


for a in twitter:
    if 'delete' in a:
        user.append(a['delete']['status']['user_id_str'])   #здесь удаленные и не удаленные надо чуть по-разному обрабатывать
    else:
        user.append(a['user']['id_str'])
counter = collections.Counter(user)  #создали словарь
values = list(counter.values())  #извлекли значения
users = 0
for a in values:
    if a > 1:
        users = users + 1
print(users, 'пользователей встречается больше одного раза')  #вывели кол-во пользователей, ЗАДАНИЕ 4


for a in twitter:
    if 'delete' in a:
        lang = lang 
    else:
        for b in a['entities']['hashtags']:
            hashtag.append(b['text'])
counter = collections.Counter(hashtag)
hashtags = []
for key in sorted(counter, key=counter.get):    #еще в прошлом году бесило, что получается так громоздко
    hashtags.append(key)                        #научите, пожалуйста, как нормально
hashtags.reverse()
print(hashtags[:20], '- топ 20 самых популярных тегов' ) #нашли теги, ЗАДАНИЕ 5


for a in twitter:
    if 'delete' in a:
        lang = lang 
    else:
        if a['lang'] == 'en':
                if a['retweeted'] == False: #чистим от ссылок, юзернеймов, хештегов, знаков препинания, приводим к н.регистру
                    s = a['text']
                    s = re.sub('RT', '', s)
                    s = re.sub('http\S+?', '', s)
                    s = re.sub('@\S+?', '', s)
                    s = re.sub('#\S+?', '', s)
                    s = re.sub('t\.co/\S+?', '', s)
                    s = s.translate(str.maketrans('', '', string.punctuation))
                    s = s.lower()
                    english.append(s)
words = []
for a in english:
    words= words + a.split()
counter = collections.Counter(words) 
words = []
for key in sorted(counter, key=counter.get):   #еще в прошлом году бесило, что получается так громоздко
    words.append(key)                          #научите, пожалуйста, как нормально
words.reverse()
print(words[:20], '- топ 20 самых популярных слов' )   #нашли слова, ЗАДАНИЕ 6


for a in twitter:
    if 'delete' in a:
        lang = lang
    else:
        followers.append([a['user']['followers_count'], a['user']['id']])  #с повторениями
followers2 = []
for a in followers:
    if a not in followers2:
        followers2.append(a)  #без повторений
df = pandas.DataFrame(followers2) #создали датафрейм, отсоритровали по кол-ву подписчиков
df.columns = ['followers', 'id']
print(df.sort_values('followers', ascending = False)[:10]) #топ 10 самых популярных пользователей, ЗАДАНИЕ 7


sources = []
for a in twitter:
    if 'delete' in a:
        lang = lang   # надо было что-нибудь вставить, решил не заморачиваться и не искать, как красиво сделать
    else:
        sources.append(a['source'])
counter = collections.Counter(sources)  #создали словарь из того, что в графе соурс
sources = []
for key in sorted(counter, key=counter.get):   #еще в прошлом году бесило, что получается так громоздко
    sources.append(key)                          #научите, пожалуйста, как нормально
sources.reverse()
clean_sources = []
for source in sources:  #из каждого грязного соурс вытаскиваем чистый
    result = re.search('>(.*?)<', source) 
    if result:
        clean_source = result.group(1)
    clean_sources.append(clean_source)
print(clean_sources[:10], '- топ 10 источников твитов' ) #вывели источники, ЗАДАНИЕ 8
```
